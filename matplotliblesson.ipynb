```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
```

## Functional Plot

```python
x_1 = np.linspace(0,5,10)
y_1 = x_1**2 
plt.plot(x_1,y_1)
plt.title('Squared growth')
```

## Multiple plots

```python
plt.subplot(1,2,1)
plt.plot(x_1, y_1, 'r')
plt.subplot(1,2,2)
plt.plot(x_1, y_1, 'b')
```

# Using Figure objects


```python
fig_1 = plt.figure(figsize=(5,4),dpi=100)
axes_1 = fig_1.add_axes([0.1,0.1,0.9,0.9])
axes_1.set_xlabel('Days')
axes_1.set_ylabel('Days squared')
axes_1.set_title('Days Squared Chart')
axes_1.plot(x_1,y_1,label='x/x**2')
axes_1.plot(y_1,x_1,label='x**2/x')
axes_1.legend(loc=0)
axes_2 = fig_1.add_axes([0.45,0.45,0.4,0.3])
axes_2.set_xlabel('Days')
axes_2.set_ylabel('Days squared')
axes_2.set_title('Days Squared Chart')
axes_2.plot(x_1,y_1, 'r')
axes_2.text(0,40, 'Message')
```

## Subplots

```python
fig_2, axes_2 = plt.subplots(figsize=(8,4), nrows=1, ncols=3)
plt.tight_layout()
axes_2[1].set_title('Plot2')
axes_2[1].set_xlabel('X')
axes_2[1].set_ylabel('X squared')
axes_2[1].plot(x_1,y_1)
```

## Options on appeareance
Default colors: b:blue, g:green, r:red, c:cyan, m:magenta, y:yellow, k:black,w:white
More stuff in cheat sheet on Sana

```python
fig_3 = plt.figure(figsize=(6,4))
axes_3 = fig_3.add_axes([0,0,1,1])
axes_3.plot(x_1,y_1, color='navy', alpha =.75, lw=2, ls='-', marker='o', markersize=7, markerfacecolor='y',markeredgecolor='y' )
axes_3.grid(True, color='0.6', dashes=(5,2,1,2))
```

## Save to file

```python
fig_3.savefig('3rd_plot.png')
```

```python
denmark_debt = gov_debt_df[gov_debt_df['country']=='Denmark']
sweden_debt = gov_debt_df[gov_debt_df['country'] == 'Sweden']
denmark_debt['debtpercentage'] = denmark_debt['debtpercentage'].str.replace(',', '.').astype(float)
sweden_debt['debtpercentage'] = sweden_debt['debtpercentage'].str.replace(',', '.').astype(float)
plt.figure(figsize=(12, 8))
plt.title('Scandinavian debt during the last 70-years')
plt.plot(denmark_debt['year'], denmark_debt['debtpercentage'], marker='o', linestyle='-', color='blue')
plt.plot(sweden_debt['year'], sweden_debt['debtpercentage'], marker='o', linestyle='-', color='red')
plt.xlabel('Year')
plt.ylabel('Debt percentage')
plt.grid(True)
plt.show()
```
