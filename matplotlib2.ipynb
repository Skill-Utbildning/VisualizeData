# Import Required Libraries
Import the necessary libraries, including Matplotlib and NumPy.

```python
# Import Required Libraries
import matplotlib.pyplot as plt
import numpy as np
```

# Basic Plotting
Create a basic line plot using Matplotlib.

```python
# Basic Plotting

# Create data for the plot
x = np.linspace(0, 10, 100)  # 100 evenly spaced values from 0 to 10
y = np.sin(x)  # Sine of x

# Create the plot
plt.figure(figsize=(10, 6))  # Set the figure size
plt.plot(x, y, label='sin(x)')  # Plot y vs. x with a label

# Add title and labels
plt.title('Basic Line Plot')
plt.xlabel('x')
plt.ylabel('sin(x)')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

# Plot Customization
Customize the plot by adding titles, labels, and changing colors.

```python
# Plot Customization

# Create data for the plot
x = np.linspace(0, 10, 100)  # 100 evenly spaced values from 0 to 10
y = np.sin(x)  # Sine of x

# Create the plot
plt.figure(figsize=(10, 6))  # Set the figure size
plt.plot(x, y, label='sin(x)', color='blue', linestyle='--')  # Plot y vs. x with a label, blue color, and dashed line

# Add title and labels
plt.title('Customized Line Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

# Add a legend
plt.legend()

# Customize the grid
plt.grid(True, which='both', linestyle=':', linewidth=0.5)

# Show the plot
plt.show()
```

# Subplots
Create multiple subplots in a single figure.

```python
# Subplots

# Create data for the subplots
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create a figure with multiple subplots
fig, axs = plt.subplots(2, 1, figsize=(10, 10))  # 2 rows, 1 column

# First subplot
axs[0].plot(x, y1, label='sin(x)', color='blue')
axs[0].set_title('Sine Function')
axs[0].set_xlabel('x')
axs[0].set_ylabel('sin(x)')
axs[0].legend()
axs[0].grid(True)

# Second subplot
axs[1].plot(x, y2, label='cos(x)', color='red')
axs[1].set_title('Cosine Function')
axs[1].set_xlabel('x')
axs[1].set_ylabel('cos(x)')
axs[1].legend()
axs[1].grid(True)

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

# Histograms
Generate and customize histograms using Matplotlib.

```python
# Histograms

# Generate random data
data = np.random.randn(1000)

# Create the histogram
plt.figure(figsize=(10, 6))
plt.hist(data, bins=30, edgecolor='black')

# Add title and labels
plt.title('Histogram of Random Data')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Show the plot
plt.show()

# Customized Histogram

# Create the histogram with customization
plt.figure(figsize=(10, 6))
plt.hist(data, bins=30, edgecolor='black', color='skyblue', alpha=0.7)

# Add title and labels
plt.title('Customized Histogram of Random Data')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Add grid
plt.grid(True, linestyle='--', linewidth=0.5)

# Show the plot
plt.show()
```

# Scatter Plots
Create and customize scatter plots.

```python
# Scatter Plots

# Generate random data for scatter plot
np.random.seed(0)
x = np.random.rand(100)
y = np.random.rand(100)
colors = np.random.rand(100)
sizes = 1000 * np.random.rand(100)

# Create the scatter plot
plt.figure(figsize=(10, 6))
scatter = plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='viridis')

# Add title and labels
plt.title('Scatter Plot with Color and Size Customization')
plt.xlabel('x')
plt.ylabel('y')

# Add color bar
plt.colorbar(scatter)

# Show the plot
plt.show()
```
